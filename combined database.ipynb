```python
def get_db():
    from pymongo import MongoClient
    
    try:
        conn = MongoClient('mongodb://192.168.2.20:27017')
        # Connect to DB "mydbname"
        db = conn.prices_db
        return db    
    except pymongo.errors.ConnectionFailure, e:
        print "Could not connect to MongoDB: %s" % e 
        raise e
```

```python
# Get the connection to our source database
mydb = get_db()

import pandas as pd

# ONS Item No 310218 is Apple Cider, 310421 is red wine, 211709 is semi-skimmed milk
df = pd.DataFrame(list(mydb.combined.find({"ons_item_no": { "$in": [310218, 310421, 211710,211709]}}, \
        {"_id":0,"timestamp":1,"ml_prediction":1,"store":1, "ons_item_no":1,"ons_item_name":1,"std_price":1, "std_unit":1,"product_name":1})))

# Show the first few records in the frame
 
```
